```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Specify the folder containing the CSV files
folder_path = "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files"

# List all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv') and not f.startswith('D4G')]

# Plot each CSV file
plt.figure(figsize=(16, 6))
for csv_file in csv_files:
    # Read the CSV file into a DataFrame
    file_path = os.path.join(folder_path, csv_file)
    data = pd.read_csv(file_path)

    # Check if the necessary columns exist
    if 'Wavelength' in data.columns and 'Reflectance' in data.columns:
        plt.plot(data['Wavelength'], data['Reflectance'], label=csv_file)

# Customize the plot
plt.title('Wavelength vs Reflectance for All CSV Files')
plt.xlabel('Wavelength')
plt.ylabel('Reflectance')
plt.legend()
plt.show()
```
