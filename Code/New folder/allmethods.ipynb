```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV files without headers
file_path_1 = "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D1GB1.csv"  # Replace with the actual path to D1GB1.csv
file_path_2 = "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D6GB1.csv"  # Replace with the actual path to D1GB2.csv


# Read the data without headers, and treat the second column as reflectance values
data1 = pd.read_csv(file_path_1, header=None)
data2 = pd.read_csv(file_path_2, header=None)

# Extract reflectance values from the second column of each file
reflectance1 = pd.to_numeric(data1.iloc[:, 1], errors='coerce').values
reflectance2 = pd.to_numeric(data2.iloc[:, 1], errors='coerce').values

# Remove any NaN values that might have resulted from non-numeric entries
reflectance1 = reflectance1[~np.isnan(reflectance1)]
reflectance2 = reflectance2[~np.isnan(reflectance2)]

# Ensure both vectors have the same length (same number of bands)
if len(reflectance1) != len(reflectance2):
    raise ValueError("The two vectors must have the same number of bands (same length).")

# 1. Spectral Angle Mapper (SAM)
numerator = np.sum(reflectance1 * reflectance2)
denominator = np.sqrt(np.sum(reflectance1 ** 2)) * np.sqrt(np.sum(reflectance2 ** 2))
angle_radians = np.arccos(numerator / denominator)
angle_degrees = np.degrees(angle_radians)
print(f"Spectral Angle (SAM) in degrees: {angle_degrees}")

# 2. Euclidean Distance
euclidean_distance = np.sqrt(np.sum((reflectance1 - reflectance2) ** 2))
print(f"Euclidean Distance: {euclidean_distance}")

# 3. Spectral Information Divergence (SID)
sid = np.sum(reflectance1 * np.log(reflectance1 / reflectance2)) + np.sum(reflectance2 * np.log(reflectance2 / reflectance1))
print(f"Spectral Information Divergence (SID): {sid}")

# 4. Normalized Difference Spectral Index (NDSI)
ndsi = (reflectance1 - reflectance2) / (reflectance1 + reflectance2)
print(f"Normalized Difference Spectral Index (NDSI): {ndsi}")

# 5. Manhattan (City Block) Distance
manhattan_distance = np.sum(np.abs(reflectance1 - reflectance2))
print(f"Manhattan Distance: {manhattan_distance}")

# 6. Cosine Similarity (alternative to SAM)
cosine_similarity = numerator / denominator
print(f"Cosine Similarity: {cosine_similarity}")

# 7. Correlation Coefficient
correlation_coefficient = np.corrcoef(reflectance1, reflectance2)[0, 1]
print(f"Correlation Coefficient: {correlation_coefficient}")

# Plot the reflectance values for both datasets for visual comparison
plt.figure(figsize=(10, 6))
plt.plot(reflectance1, label='Dataset 1 Reflectance', marker='o')
plt.plot(reflectance2, label='Dataset 2 Reflectance', marker='x')
plt.xlabel('Band Index')
plt.ylabel('Reflectance')
plt.title('Reflectance Comparison between Dataset 1 and Dataset 2')
plt.legend()
plt.grid(True)
plt.show()
```
