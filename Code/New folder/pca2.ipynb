```python
import numpy as np
import pandas as pd
import matplotlib as plt
from numpy.linalg import eig
```

```python
file_path ="output_reflectance4.csv"

df = pd.read_csv(file_path)

file_names = df.iloc[:, 0].values

feature_names = df.columns[1:]

data = df.iloc[:, 1:].values

print("File Names:", file_names)
print("Feature Names:", feature_names)
print("Data as NumPy Array:\n", data)
```

```python
print(df)
```

```python
numeric_data = df.iloc[:, 1:].values
print(numeric_data)
```

```python
mean_centered_data = numeric_data - np.mean(numeric_data, axis=0)
normalized_data = mean_centered_data / np.std(numeric_data, axis=0)
```

```python
print(normalized_data)
```

```python
cov_matrix = np.cov(normalized_data, rowvar=False)
print(cov_matrix)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```
