```python
df1 = pd.read_csv("PSR-1100_SN12B4415_M0632.csv")

df2 = pd.read_csv("PSR-1100_SN12B4415_M0796.csv")

# Check the columns in the DataFrames to ensure correct column names are used
print(df1.columns)
print(df2.columns)

# Plot the data from both CSV files
plt.figure(figsize=(10, 6))

# Plot data from the first file
plt.plot(df1[df1.columns[0]], df1[df1.columns[1]], label='Crocin650: Reflectance', color='blue')

# Plot data from the second file
plt.plot(df2[df2.columns[0]], df2[df2.columns[1]], label='counterfit Durg: Reflectance', color='red')

# Add titles and labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (%)')
plt.title('Reflectance Spectrum Comparison')

# Add a legend
plt.legend()

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()
```

```python
# prompt: where value is change show in the graph using derivative

import numpy as np

# Calculate the derivative of the reflectance data for both dataframes
df1_derivative = np.gradient(df1[df1.columns[1]], df1[df1.columns[0]])
df2_derivative = np.gradient(df2[df2.columns[1]], df2[df2.columns[0]])

# Plot the derivative data
plt.figure(figsize=(10, 6))


plt.plot(df1[df1.columns[0]], df1_derivative, label='Crocin650: Derivative of Reflectance', color='blue')


plt.plot(df2[df2.columns[0]], df2_derivative, label='counterfit Durg: Derivative of Reflectance', color='red')


plt.xlabel('Wavelength (nm)')
plt.ylabel('Derivative of Reflectance')
plt.title('Derivative of Reflectance Spectrum Comparison')

# Add a legend
plt.legend()

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()
```

```python


# Calculate the difference between the two reflectance spectra
difference = df1[df1.columns[1]] - df2[df2.columns[1]]

# Plot the difference spectrum
plt.figure(figsize=(10, 6))
plt.plot(df1[df1.columns[0]], difference, label='Difference in Reflectance', color='purple')

# Add titles and labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Difference in Reflectance (%)')
plt.title('Difference in Reflectance Spectrum')

# Add a legend
plt.legend()

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()

# Calculate the derivative of the difference spectrum
difference_derivative = np.gradient(difference, df1[df1.columns[0]])

# Plot the derivative of the difference spectrum
plt.figure(figsize=(10, 6))
plt.plot(df1[df1.columns[0]], difference_derivative, label='Derivative of Difference in Reflectance', color='purple')

# Add titles and labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Derivative of Difference in Reflectance')
plt.title('Derivative of Difference in Reflectance Spectrum')

# Add a legend
plt.legend()

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()
```

```python
# prompt: in 500nm how much differece compare both graph

# Find the index of the closest wavelength to 500nm in both dataframes
index1 = np.argmin(np.abs(df1[df1.columns[0]] - 500))
index2 = np.argmin(np.abs(df2[df2.columns[0]] - 500))

# Calculate the difference in reflectance at 500nm
difference_at_500 = df1[df1.columns[1]][index1] - df2[df2.columns[1]][index2]

# Print the difference
print(f"Difference in reflectance at 500nm: {difference_at_500:.2f}%")
```

```python
# prompt: All value of wavelength difference compare both graph and visualize

# Calculate the wavelength difference between the two dataframes
wavelength_difference = df1[df1.columns[0]] - df2[df2.columns[0]]

# Plot the wavelength difference
plt.figure(figsize=(10, 6))
plt.plot(df1[df1.columns[0]], wavelength_difference, label='Wavelength Difference', color='purple')

# Add titles and labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Wavelength Difference (nm)')
plt.title('Wavelength Difference between Spectra')

# Add a legend
plt.legend()

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()
```
