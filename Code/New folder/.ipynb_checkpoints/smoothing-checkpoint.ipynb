```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV
data = pd.read_csv("C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D1GB1.csv")
wavelength = data['Wavelength'].values
reflectance = data['Reflectance'].values

# Coefficients for weighted smoothing (adjust if needed)
C = np.array([1, 2, 3, 2, 1])  # Example weights
m = (len(C) - 1) // 2  # Half-window size
N = np.sum(C)  # Normalization factor

# Calculate smoothed reflectance values
reflectance_smooth = np.array([
    np.sum(C * reflectance[j - m : j + m + 1]) / N
    for j in range(m, len(reflectance) - m)
])

# Adjust wavelength to match smoothed reflectance length
wavelength_smooth = wavelength[m:len(wavelength) - m]

# Plot original and smoothed data
plt.figure(figsize=(14, 6))
plt.plot(wavelength, reflectance, label="Original Reflectance", marker='o', linestyle='-', color='blue')
plt.plot(wavelength_smooth, reflectance_smooth, label="Smoothed Reflectance", marker='o', linestyle='-', color='red')
plt.xlabel("Wavelength")
plt.ylabel("Reflectance")
plt.title("Reflectance vs Wavelength (Original and Smoothed)")
plt.legend()
plt.grid(True)
plt.show()
```
