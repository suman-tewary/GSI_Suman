```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Define the directory path where your CSV files are located
directory ="C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files"

# Define the target wavelengths for Green and Red
green_target = 550  # Green wavelength (typically around 550 nm)
red_target = 660  # Red wavelength (typically around 660 nm)

# Function to calculate reflectance from a CSV file at a given target wavelength
def calculate_reflectance(file_path, target_wavelength):
    # Load the data from the CSV file
    data = pd.read_csv(file_path)

    # Find the row with the closest wavelength to the target (Green or Red)
    target_row = data.iloc[(data['Wavelength'] - target_wavelength).abs().argmin()]

    # Extract the reflectance value for the target wavelength
    target_reflectance = target_row['Reflectance']
    
    return target_reflectance

# Lists to store reflectance values for each day (Green and Red)
green_reflectance = []
red_reflectance = []
days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5', 'Day 6']

# Loop through all the days and process files
for day in range(1, 7):
    # Create filenames based on the current day
    green_file_name = f"D{day}GT.csv"  # Files with 'GT' in the name (Green)
    red_file_name = f"D{day}RT.csv"   # Files with 'RT' in the name (Red)
    
    # Find the full path to the Green and Red files (considering the possibility of T1, S1, etc.)
    green_file_path = None
    red_file_path = None
    
    for file_name in os.listdir(directory):
        if green_file_name[:-4] in file_name and 'GT' in file_name:  # Match 'GT' for Green
            green_file_path = os.path.join(directory, file_name)
        if red_file_name[:-4] in file_name and 'RT' in file_name:  # Match 'RT' for Red
            red_file_path = os.path.join(directory, file_name)

    # Calculate Green reflectance if the file exists
    if green_file_path:
        green_ref = calculate_reflectance(green_file_path, green_target)
        green_reflectance.append(green_ref)
    else:
        green_reflectance.append(None)

    # Calculate Red reflectance if the file exists
    if red_file_path:
        red_ref = calculate_reflectance(red_file_path, red_target)
        red_reflectance.append(red_ref)
    else:
        red_reflectance.append(None)

# Calculate NDVI for each day using the formula:
# NDVI = (Green - Red) / (Green + Red)
ndvi_values = [(green - red) / (green + red) if green is not None and red is not None else None
               for green, red in zip(green_reflectance, red_reflectance)]

# Plot Green Reflectance vs Days
plt.figure(figsize=(10, 6))
plt.plot(days, green_reflectance, marker='o', linestyle='-', color='g', label='Green Reflectance')
plt.title('Green Reflectance for Each Day')
plt.xlabel('Days')
plt.ylabel('Reflectance')
plt.grid(True)
plt.legend()
plt.show()

# Plot Red Reflectance vs Days
plt.figure(figsize=(10, 6))
plt.plot(days, red_reflectance, marker='o', linestyle='-', color='r', label='Red Reflectance')
plt.title('Red Reflectance for Each Day')
plt.xlabel('Days')
plt.ylabel('Reflectance')
plt.grid(True)
plt.legend()
plt.show()

# Plot NDVI vs Days
plt.figure(figsize=(10, 6))
plt.plot(days, ndvi_values, marker='o', linestyle='-', color='b', label='NDVI')
plt.title('NDVI for Each Day')
plt.xlabel('Days')
plt.ylabel('NDVI')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D1RT1.csv"
data = pd.read_csv(file_path)

# Define the target wavelengths for each band
target_wavelengths = {
    'Green': 550,
    'Red': 660,
    'NIR': 800,
    'SWIR': 1600
}

# Find the closest available wavelength in the data for each target band
band_values = {}
for band, target in target_wavelengths.items():
    # Find the row with the wavelength closest to the target
    closest_row = data.iloc[(data['Wavelength'] - target).abs().argsort()[:1]]
    # Store the reflectance value for that band
    band_values[band] = closest_row['Reflectance'].values[0]

# Calculate spectral indices
# NDVI: (NIR - Red) / (NIR + Red)
ndvi = (band_values['NIR'] - band_values['Red']) / (band_values['NIR'] + band_values['Red'])

# ARI: (1 / Green) - (1 / Red)
ari = (1 / band_values['Green']) - (1 / band_values['Red'])

# MSI: NIR / SWIR
msi = band_values['NIR'] / band_values['SWIR']

# REP (Red-Edge Position): Approximate as average of NIR and Red
rep = (band_values['NIR'] + band_values['Red']) / 2

# Display the results
print("Spectral Indices Calculated:")
print(f"NDVI: {ndvi}")
print(f"ARI: {ari}")
print(f"MSI: {msi}")
print(f"REP: {rep}")

```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Manually specify the paths to the 6 CSV files
csv_files = [
     "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D1GS1.csv",
     "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D2GS1.csv",
     "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D3GS1.csv",
     # "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D4GS1.csv",
     "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D5GS1.csv",
     "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D6GS1.csv"
]

# Define the target wavelengths for each band
target_wavelengths = {
    'Green': 550,
    'Red': 660,
    'NIR': 800,
    'SWIR': 1600
}

# Lists to store index values and file names for plotting
file_names = []
ndvi_values = []
ari_values = []
msi_values = []
rep_values = []

# Process each of the 6 specified CSV files
for file_path in csv_files:
    if os.path.exists(file_path):  # Check if the file exists
        # Load the CSV file
        data = pd.read_csv(file_path)
        
        # Find the closest available wavelength in the data for each target band
        band_values = {}
        for band, target in target_wavelengths.items():
            closest_row = data.iloc[(data['Wavelength'] - target).abs().argsort()[:1]]
            band_values[band] = closest_row['Reflectance'].values[0]
        
        # Calculate spectral indices
        ndvi = (band_values['NIR'] - band_values['Red']) / (band_values['NIR'] + band_values['Red'])
        ari = (1 / band_values['Green']) - (1 / band_values['Red'])
        msi = band_values['NIR'] / band_values['SWIR']
        rep = (band_values['NIR'] + band_values['Red']) / 2

        # Append calculated values and filename to lists
        file_names.append(os.path.basename(file_path))
        ndvi_values.append(ndvi)
        ari_values.append(ari)
        msi_values.append(msi)
        rep_values.append(rep)

# Plot each spectral index in a separate graph with values displayed
plt.figure(figsize=(12, 10))

# NDVI plot
plt.subplot(2, 2, 1)
plt.plot(file_names, ndvi_values, marker='o', color='b', label='NDVI')
for i, value in enumerate(ndvi_values):
    plt.text(file_names[i], value, f"{value:.2f}", ha='center', va='bottom')
plt.xticks(rotation=45, ha='right')
plt.xlabel('File')
plt.ylabel('NDVI')
plt.title('NDVI for Each File')
plt.legend()

# ARI plot
plt.subplot(2, 2, 2)
plt.plot(file_names, ari_values, marker='o', color='r', label='ARI')
for i, value in enumerate(ari_values):
    plt.text(file_names[i], value, f"{value:.2f}", ha='center', va='bottom')
plt.xticks(rotation=45, ha='right')
plt.xlabel('File')
plt.ylabel('ARI')
plt.title('ARI for Each File')
plt.legend()

# MSI plot
plt.subplot(2, 2, 3)
plt.plot(file_names, msi_values, marker='o', color='g', label='MSI')
for i, value in enumerate(msi_values):
    plt.text(file_names[i], value, f"{value:.2f}", ha='center', va='bottom')
plt.xticks(rotation=45, ha='right')
plt.xlabel('File')
plt.ylabel('MSI')
plt.title('MSI for Each File')
plt.legend()

# REP plot
plt.subplot(2, 2, 4)
plt.plot(file_names, rep_values, marker='o', color='m', label='REP')
for i, value in enumerate(rep_values):
    plt.text(file_names[i], value, f"{value:.2f}", ha='center', va='bottom')
plt.xticks(rotation=45, ha='right')
plt.xlabel('File')
plt.ylabel('REP')
plt.title('REP for Each File')
plt.legend()

plt.tight_layout()
plt.show()

```
