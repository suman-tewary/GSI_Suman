# First order derivative

```python
import numpy as np
import matplotlib.pyplot as plt

# Load your data
late_mature_apple = np.loadtxt('/content/drive/MyDrive/Dataset_of_Apple/mature_apple (1).txt')
mature_apple = np.loadtxt('/content/drive/MyDrive/Dataset_of_Apple/late_mature_apple (1).txt')

# Calculate first order derivatives
 # dfdx = np.gradient (fx, dx), where fx = Reflectance values and dx= Wabelength values
late_mature_apple_first_derivative = np.gradient(late_mature_apple[:, 1], late_mature_apple[:, 0])

 # dfdx = np.gradient (fx, dx), where fx = Reflectance values and dx= Wabelength values
mature_apple_first_derivative = np.gradient(mature_apple[:, 1], mature_apple[:, 0])


# Plot the first order derivatives
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1) #plt.subplot (nrows, ncols, index)
plt.plot(late_mature_apple[:, 0], late_mature_apple_first_derivative, label='Late Mature Apple', color='blue')
plt.xlabel('Wavelength (nm)')
plt.ylabel('First Derivative of Reflectance')
plt.title('First Order Derivative for Late Mature Apple')
plt.legend()

plt.subplot(1, 2, 2) #plt.subplot (nrows, ncols, index)
plt.plot(mature_apple[:, 0], mature_apple_first_derivative, label='Mature Apple', color='green')
plt.xlabel('Wavelength (nm)')
plt.ylabel('First Derivative of Reflectance')
plt.title('First Order Derivative for Mature Apple')
plt.legend()

plt.tight_layout()
plt.show()
```

# Differenciate by using SAM(Spectral Angle Mapper)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

# Load the data for "Mature Apple"
mature_apple = '/content/drive/MyDrive/Dataset_of_Apple/mature_apple (1).txt'
mature_df = pd.read_csv(mature_apple, delim_whitespace=True, names=['Wavelength', 'Reflectance'])

# Load the data for "Late Mature Apple"
late_mature_apple = '/content/drive/MyDrive/Dataset_of_Apple/late_mature_apple (1).txt'
late_mature_df = pd.read_csv(late_mature_apple, delim_whitespace=True, names=['Wavelength', 'Reflectance'])

# Make sure the Wavelength values are the same for both datasets
# If not, you may need to interpolate to match wavelengths between the two datasets

# Define the SAM function
def spectral_angle_mapper(spectrum1, spectrum2):
    dot_product = np.dot(spectrum1, spectrum2)
    magnitude1 = np.linalg.norm(spectrum1)
    magnitude2 = np.linalg.norm(spectrum2)
    angle = math.acos(dot_product / (magnitude1 * magnitude2))
    return angle

# Calculate SAM for Reflectance values
sam_value = spectral_angle_mapper(mature_df['Reflectance'], late_mature_df['Reflectance'])

# Plotting both spectra for comparison
plt.figure(figsize=(8, 4))

# Plot Mature Apple
plt.plot(mature_df['Wavelength'], mature_df['Reflectance'], label='Mature Apple', color='green')

# Plot Late Mature Apple
plt.plot(late_mature_df['Wavelength'], late_mature_df['Reflectance'], label='Late Mature Apple', color='blue')

# Add labels and title
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance')
plt.title(f'Spectral Comparison (SAM Angle = {sam_value:.4f} radians)')
plt.legend()
plt.grid(True)
plt.show()
```

# By using Distance Function

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

# Load the data for "Mature Apple"
mature_apple = '/content/drive/MyDrive/Dataset_of_Apple/mature_apple (1).txt'
mature_df = pd.read_csv(mature_apple, delim_whitespace=True, names=['Wavelength', 'Reflectance'])

# Load the data for "Late Mature Apple"
late_mature_apple = '/content/drive/MyDrive/Dataset_of_Apple/late_mature_apple (1).txt'
late_mature_df = pd.read_csv(late_mature_apple, delim_whitespace=True, names=['Wavelength', 'Reflectance'])

# Define the SAM function
def spectral_angle_mapper(spectrum1, spectrum2):
    dot_product = np.dot(spectrum1, spectrum2)
    magnitude1 = np.linalg.norm(spectrum1)
    magnitude2 = np.linalg.norm(spectrum2)
    angle = math.acos(dot_product / (magnitude1 * magnitude2))
    return angle

# Calculate SAM for Reflectance values as initial distance measure d_k
sam_value = spectral_angle_mapper(mature_df['Reflectance'], late_mature_df['Reflectance'])

# Generate a series of eta values (parameter values) for demonstration
eta_values = np.linspace(0, 1, 10)  # Assuming eta values range from 0 to 1
sam_distances = np.array([spectral_angle_mapper(mature_df['Reflectance'], late_mature_df['Reflectance'] * (1 + e)) for e in eta_values])

# Equation (1): Normalize and apply inequality
eta_min, eta_max = eta_values[0], eta_values[-1]
dk_min, dk_max = sam_distances[0], sam_distances[-1]

normalized_distances = (sam_distances - dk_min) / (dk_max - dk_min)
eta_ratios = (eta_values - eta_min) / (eta_max - eta_min)
inequality_satisfied = np.all(normalized_distances >= eta_ratios)

print("Inequality (1) satisfied for all eta values:" if inequality_satisfied else "Inequality (1) not satisfied for all eta values.")

# Equation (2): Calculate eta_sat where distance function reaches its maximum
eta_sat = eta_values[np.argmax(sam_distances)]
print("Saturation point (eta_sat):", eta_sat)

# Equation (3): Calculate dynamic range dnr
dnr = dk_max - dk_min
print("Dynamic range (dnr):", dnr)

# Plotting the distance function over eta values
plt.figure(figsize=(8, 4))
plt.plot(eta_values, sam_distances, label='Distance function d_k(eta)', color='purple')
plt.xlabel('Eta')
plt.ylabel('Distance (SAM)')
plt.title(f'Distance Function with Dynamic Range (dnr = {dnr:.4f})')
plt.axvline(x=eta_sat, color='red', linestyle='--', label=f'Saturation Point (eta_sat = {eta_sat:.2f})')
plt.legend()
plt.grid(True)
plt.show()
```

# GMC Distance Function

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data for "Mature Apple"
mature_apple = '/content/drive/MyDrive/Dataset_of_Apple/mature_apple (1).txt'
mature_df = pd.read_csv(mature_apple, delim_whitespace=True, names=['Wavelength', 'Reflectance'])

# Load the data for "Late Mature Apple"
late_mature_apple = '/content/drive/MyDrive/Dataset_of_Apple/late_mature_apple (1).txt'
late_mature_df = pd.read_csv(late_mature_apple, delim_whitespace=True, names=['Wavelength', 'Reflectance'])

# Define the GMC distance function
def gmc_distance(spectrum1, spectrum2):
    differences = spectrum1 - spectrum2
    gmc_values = (differences ** 2) / (1 + differences ** 2)
    return np.sum(gmc_values)

# Calculate GMC for Reflectance values as initial distance measure d_k
gmc_value = gmc_distance(mature_df['Reflectance'].values, late_mature_df['Reflectance'].values)

# Generate a series of eta values (parameter values) for demonstration
eta_values = np.linspace(0, 1, 10)  # Assuming eta values range from 0 to 1
gmc_distances = np.array([gmc_distance(mature_df['Reflectance'].values, late_mature_df['Reflectance'].values * (1 + e)) for e in eta_values])

# Equation (1): Normalize and apply inequality
eta_min, eta_max = eta_values[0], eta_values[-1]
dk_min, dk_max = gmc_distances[0], gmc_distances[-1]

normalized_distances = (gmc_distances - dk_min) / (dk_max - dk_min)
eta_ratios = (eta_values - eta_min) / (eta_max - eta_min)
inequality_satisfied = np.all(normalized_distances >= eta_ratios)

print("Inequality (1) satisfied for all eta values:" if inequality_satisfied else "Inequality (1) not satisfied for all eta values.")

# Equation (2): Calculate eta_sat where distance function reaches its maximum
eta_sat = eta_values[np.argmax(gmc_distances)]
print("Saturation point (eta_sat):", eta_sat)

# Equation (3): Calculate dynamic range dnr
dnr = dk_max - dk_min
print("Dynamic range (dnr):", dnr)

# Plotting the distance function over eta values
plt.figure(figsize=(8, 4))
plt.plot(eta_values, gmc_distances, label='Distance function d_k(eta)', color='purple')
plt.xlabel('Eta')
plt.ylabel('Distance (GMC)')
plt.title(f'Distance Function with Dynamic Range (dnr = {dnr:.4f})')
plt.axvline(x=eta_sat, color='red', linestyle='--', label=f'Saturation Point (eta_sat = {eta_sat:.2f})')
plt.legend()
plt.grid(True)
plt.show()
```
