```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV files without headers
file_path_1 = "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D1GB1.csv" 
file_path_2 = "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D3GB1.csv" 


# Read the data without headers, and treat the first column as reflectance values
data1 = pd.read_csv(file_path_1, header=None)
data2 = pd.read_csv(file_path_2, header=None)

# Convert the first column to numeric values to avoid string issues
reflectance1 = pd.to_numeric(data1.iloc[:, 0], errors='coerce').values
reflectance2 = pd.to_numeric(data2.iloc[:, 0], errors='coerce').values

# Drop any NaN values that might have resulted from non-numeric entries
reflectance1 = reflectance1[~np.isnan(reflectance1)]
reflectance2 = reflectance2[~np.isnan(reflectance2)]

# Ensure both vectors have the same length (same number of bands)
if len(reflectance1) != len(reflectance2):
    raise ValueError("The two vectors must have the same number of bands (same length).")

# SAM formula implementation
numerator = np.sum(reflectance1 * reflectance2)
denominator = np.sqrt(np.sum(reflectance1 ** 2)) * np.sqrt(np.sum(reflectance2 ** 2))
angle_radians = np.arccos(numerator / denominator)

# Convert angle to degrees for easier interpretation
angle_degrees = np.degrees(angle_radians)
print(f"Spectral Angle (in degrees): {angle_degrees}")

# Plot the reflectance values for both datasets
plt.figure(figsize=(10, 6))
plt.plot(reflectance1, label='Dataset 1 Reflectance', marker='o')
plt.plot(reflectance2, label='Dataset 2 Reflectance', marker='x')
plt.xlabel('Band Index')
plt.ylabel('Reflectance')
plt.title('Reflectance Comparison between Dataset 1 and Dataset 2')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Day 1 - Day 2
import numpy as np
import pandas as pd

# Load the CSV data
# Replace 'spectrum1.csv' and 'spectrum2.csv' with the actual filenames
file_path_1 = "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D1GB1.csv" 
file_path_2 = "C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D2GB1.csv" 
data1 = pd.read_csv(file_path_1)
data2 = pd.read_csv(file_path_2)

# Merge data on the 'Wavelength' column to align the reflectance values
merged_data = pd.merge(data1, data2, on='Wavelength', suffixes=('_1', '_2'))

# Extract reflectance columns
reflectance1 = merged_data['Reflectance_1'].values
reflectance2 = merged_data['Reflectance_2'].values

# Calculate the SAM angle
numerator = np.sum(reflectance1 * reflectance2)
denominator = np.sqrt(np.sum(reflectance1 ** 2)) * np.sqrt(np.sum(reflectance2 ** 2))
angle_radians = np.arccos(numerator / denominator)

# Optionally, convert radians to degrees
angle_degrees = np.degrees(angle_radians)

print(f"Spectral Angle Mapper (SAM) Difference in Radians: {angle_radians}")
print(f"Spectral Angle Mapper (SAM) Difference in Degrees: {angle_degrees}")
```
