```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Read the data from CSV files
df1 = pd.read_csv("PSR-1100_SN12B4415_M0632.csv")
df2 = pd.read_csv("PSR-1100_SN12B4415_M0796.csv")

# Clean column names
df1.columns = [col.strip() for col in df1.columns]
df2.columns = [col.strip() for col in df2.columns]

# Verify column names
print("df1 columns:", df1.columns)
print("df2 columns:", df2.columns)
```

```python
# Plotting reflectance spectrum for Crocin 650
plt.figure(figsize=(10, 6))
plt.scatter(df1['Wvl'], df1['Reflect. %'], label='Reflectance: Crocin 650', color='red')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (%)')
plt.title('Reflectance Spectrum of Crocin 650')
plt.legend()
plt.grid(True)
plt.show()

# Plotting reflectance spectrum for Parasafe
plt.figure(figsize=(10, 6))
plt.scatter(df2['Wvl'], df2['Reflect. %'], label='Reflectance: Parasafe', color='blue')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (%)')
plt.title('Reflectance Spectrum of Parasafe')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the gradient (derivative) of the reflectance data
df1_derivative = np.gradient(df1['Reflect. %'], df1['Wvl'])
df2_derivative = np.gradient(df2['Reflect. %'], df2['Wvl'])

# Plot the derivative of the reflectance data
plt.figure(figsize=(10, 6))
plt.plot(df1['Wvl'], df1_derivative, label='Crocin650: Derivative of Reflectance', color='blue')
plt.plot(df2['Wvl'], df2_derivative, label='Parasafe: Derivative of Reflectance', color='red')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Derivative of Reflectance (%)')
plt.title('Derivative of Reflectance Spectrum Comparison')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Calculate RMSE and MAE for reflectance
rmse_reflectance = np.sqrt(mean_squared_error(df1['Reflect. %'], df2['Reflect. %']))
mae_reflectance = mean_absolute_error(df1['Reflect. %'], df2['Reflect. %'])

# Calculate RMSE and MAE for derivatives
rmse_derivative = np.sqrt(mean_squared_error(df1_derivative, df2_derivative))
mae_derivative = mean_absolute_error(df1_derivative, df2_derivative)

# Print the results
print(f"RMSE (Reflectance): {rmse_reflectance}")
print(f"MAE (Reflectance): {mae_reflectance}")
print(f"RMSE (Derivative): {rmse_derivative}")
print(f"MAE (Derivative): {mae_derivative}")
```

```python
# Calculate the absolute differences between reflectance and derivatives
abs_diff_reflectance = np.abs(df1['Reflect. %'] - df2['Reflect. %'])
abs_diff_derivative = np.abs(df1_derivative - df2_derivative)

# Plot the absolute differences for reflectance
plt.figure(figsize=(10, 6))
plt.plot(df1['Wvl'], abs_diff_reflectance, label='Absolute Difference: Reflectance', color='green')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absolute Difference in Reflectance (%)')
plt.title('Absolute Difference in Reflectance Spectrum')
plt.legend()
plt.grid(True)
plt.show()

# Plot the absolute differences for derivatives
plt.figure(figsize=(10, 6))
plt.plot(df1['Wvl'], abs_diff_derivative, label='Absolute Difference: Derivative of Reflectance', color='purple')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absolute Difference in Derivative of Reflectance (%)')
plt.title('Absolute Difference in Derivative of Reflectance Spectrum')
plt.legend()
plt.grid(True)
plt.show()
```
