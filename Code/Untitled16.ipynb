```python
import matplotlib.pyplot as plt

# Data
categories = ['Authentic Prozac', 'Counterfeit Samples', 'Generic Fluoxetine']
correctly_identified = [98, 90, 94]
incorrectly_identified = [2, 10, 6]

# Plotting the data
fig, ax = plt.subplots()
bar_width = 0.35
index = range(len(categories))

bar1 = plt.bar(index, correctly_identified, bar_width, label='Correctly Identified', color='b')
bar2 = plt.bar([i + bar_width for i in index], incorrectly_identified, bar_width, label='Incorrectly Identified', color='r')

plt.xlabel('Sample Types')
plt.ylabel('Percentage')
plt.title('NIR Spectroscopy Detection Accuracy')
plt.xticks([i + bar_width / 2 for i in index], categories)
plt.legend()

plt.tight_layout()
plt.show()
```
