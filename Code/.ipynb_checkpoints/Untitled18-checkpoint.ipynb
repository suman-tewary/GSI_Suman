```python
import pandas as pd
import numpy as np

# Load the data from the CSV file
df = pd.read_csv('Arsanate.csv')
reflectance_values = df['Reflectance'].values

# Step 1: Calculate Threshold T0
Nb = len(reflectance_values)  # Number of spectral bands
rho = 0.75  # Adaptive global coefficient, assuming 0.75 as given in the example
average_reflectance = np.mean(reflectance_values)
T0 = rho * average_reflectance

# Step 2: Calculate Threshold Tl
# max(y_i) only considers values greater than T0
k = np.sum(reflectance_values >= T0)
max_y = np.max(np.where(reflectance_values >= T0, reflectance_values, 0))
Tl = max_y / k

# Step 3: Calculate Threshold Tr
# min(y_i) only considers values less than T0
p = np.sum(reflectance_values < T0)
min_y = np.min(np.where(reflectance_values < T0, reflectance_values, np.inf))
Tr = min_y / p

# Step 4: Encode reflectance values into DNA codewords
def encode_to_dna(y, Tl, T0, Tr):
    if y < Tl:
        return 'T'
    elif Tl <= y < T0:
        return 'C'
    elif T0 <= y < Tr:
        return 'A'
    else:
        return 'G'

dna_codewords = [encode_to_dna(y, Tl, T0, Tr) for y in reflectance_values]

# Step 5: Save the encoded DNA codewords back into the DataFrame
df['DNA Codeword'] = dna_codewords

# Output the first few rows to verify
print(df.head())

# Save the encoded data to a new CSV file
df.to_csv('tap_water_encoded.csv', index=False)
```
