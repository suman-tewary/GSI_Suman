```python
import numpy as np
import pandas as pd
import matplotlib as plt
from numpy.linalg import eig
file_path ="output_reflectance4.csv"

df = pd.read_csv(file_path, header=0)

feature_names = df.columns[1:] 
data = df.iloc[18:25, 1:].values.astype(float)
print(data.shape)
df1=pd.DataFrame(data)
```

```python
correlation_matrix = df1.corr()
print(correlation_matrix)
```

```python
df = pd.DataFrame(data)

data_centered = df1 - df1.mean()

cov_matrix = np.cov(data_centered, rowvar=False)

print(cov_matrix)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print(eigenvalues.shape)
print(eigenvectors.shape)
```

```python
band_variances = np.diag(cov_matrix)
print(band_variances.size)
print(band_variances)
```

```python
eigenvectors=np.real(eigenvectors)
eigenvalues=np.real(eigenvalues)
factor_loadings = np.zeros_like(eigenvectors)
for k in range(eigenvectors.shape[0]): 
    for p in range(eigenvectors.shape[1]): 
        factor_loadings[k, p] = (eigenvectors[k, p] * np.sqrt(eigenvalues[p])) / np.sqrt(band_variances[k])
```

```python
first_4_factor_loadings = factor_loadings[:, :4]

print( first_4_factor_loadings)
```

```python
variance_percentage = (eigenvalues / np.sum(eigenvalues)) * 100
print(variance_percentage)
```

```python
formatted_data = np.round(variance_percentage, 3)
print(formatted_data)
```

```python
idx = eigenvalues.argsort()[::-1]

# Sort the eigenvalues in descending order 
eigenvalues = eigenvalues[idx]

# sort the corresponding eigenvectors accordingly
eigenvectors = eigenvectors[:,idx]
```

```python
explained_var = np.cumsum(eigenvalues) / np.sum(eigenvalues)
explained_var
```

```python
n_components = np.argmax(explained_var >= 0.99) + 1
n_components
```
