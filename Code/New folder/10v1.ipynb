```python
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = "C:\\Users\\sumat\\Downloads\\RT1_moc_resamp.sig"

# Read the data while skipping metadata lines
df = pd.read_csv(file_path, delim_whitespace=True, comment='/', skiprows=4, header=None)

# Assuming columns: Wavelength, (Other columns), Reflectance
df = df.iloc[:, [0, -1]]  # Selecting only Wavelength and Reflectance
df.columns = ["Wavelength", "Reflectance"]

# Plot Reflectance vs Wavelength (Switched Axes)
plt.figure(figsize=(10, 5))
plt.plot(df["Reflectance"], df["Wavelength"], label="Reflectance Spectrum", color='r')

# Customize plot
plt.title("Reflectance vs Wavelength")
plt.xlabel("Reflectance")
plt.ylabel("Wavelength (nm)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths
file1 = "C:\\Users\\sumat\\Downloads\\RT1_moc_resamp.sig"
file2 = "C:\\Users\\sumat\\Desktop\\GSI Suman\\Red Apple -By Suman\\31_07_24\\red Top 1\\RT1_moc_resamp.sig"

# Function to read and process the SIG file
def read_sig_file(file_path):
    df = pd.read_csv(file_path, delim_whitespace=True, comment='/', skiprows=4, header=None)
    df = df.iloc[:, [0, -1]]  # Selecting only Wavelength and Reflectance
    df.columns = ["Wavelength", "Reflectance"]
    return df

# Read both files
df1 = read_sig_file(file1)
df2 = read_sig_file(file2)

# Plot both datasets
plt.figure(figsize=(10, 5))
plt.plot(df1["Wavelength"], df1["Reflectance"], label="File 1", color='b')
plt.plot(df2["Wavelength"], df2["Reflectance"], label="File 2", color='r')

# Customize plot
plt.title("Wavelength vs Reflectance for Two Files")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Reflectance")
plt.legend()
plt.grid(True)
plt.show()
```
