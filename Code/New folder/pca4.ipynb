```python
import numpy as np
import pandas as pd
import matplotlib as plt
from numpy.linalg import eig
```

```python
file_path ="output_reflectance4.csv"

df = pd.read_csv(file_path, header=0)

feature_names = df.columns[1:] 
data = df.iloc[:9, 1:].values.astype(float)
print(data.shape)
```

```python
df1=pd.DataFrame(data)
```

```python

correlation_matrix = df1.corr()
print(correlation_matrix)
```

```python

df = pd.DataFrame(data)

data_centered = df1 - df1.mean()

cov_matrix = np.cov(data_centered, rowvar=False)

print(cov_matrix)

```

```python
print(cov_matrix.shape)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print(eigenvalues.shape)
print(eigenvectors.shape)
```

```python
transposed=eigenvectors.T
eig_=eigenvectors @ cov_matrix @transposed
print(eig_)
```

```python
eigenvalues = np.real(eigenvalues)
print(eigenvalues)
```

```python
variance_percentage = (eigenvalues / np.sum(eigenvalues)) * 100
print(variance_percentage)
```

```python
print(sum(variance_percentage))
```

```python
variance_percentage_df=pd.DataFrame({
    "Feature": feature_names,
    "Explained Variance (%)": variance_percentage
})
```

```python
results_sorted = variance_percentage_df.sort_values(by="Explained Variance (%)", ascending=False)
print(variance_percentage_df)
```

```python
results_sorted["Explained Variance (%)"] = results_sorted["Explained Variance (%)"].map("{:.2f}".format)
print(results_sorted)
```

```python
band_variances = np.diag(cov_matrix)
print(band_variances.size)
print(band_variances)
```

```python
eigenvectors=np.real(eigenvectors)
eigenvalues=np.real(eigenvalues)
factor_loadings = np.zeros_like(eigenvectors)
for k in range(eigenvectors.shape[0]): 
    for p in range(eigenvectors.shape[1]): 
        factor_loadings[k, p] = (eigenvectors[k, p] * np.sqrt(eigenvalues[p])) / np.sqrt(band_variances[k])
```

```python
print(factor_loadings)
```

```python
factor_loadings_df = pd.DataFrame(
    factor_loadings,
    index=[f"Band{k+1}" for k in range(eigenvectors.shape[0])],
    columns=[f"PC{p+1}" for p in range(eigenvectors.shape[1])]
)
print("Factor Loadings (R_{kp}):")
print(factor_loadings_df)
```

```python
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
top_3_eigenvalues = eigenvalues[:3]
top_3_eigenvectors = eigenvectors[:, :3]
factor_loadings = np.zeros_like(top_3_eigenvectors)
for k in range(top_3_eigenvectors.shape[0]):  # Iterate over bands
    for p in range(top_3_eigenvectors.shape[1]):  # Iterate over top 3 components
        factor_loadings[k, p] = (
            (top_3_eigenvectors[k, p] * np.sqrt(top_3_eigenvalues[p])) / np.sqrt(band_variances[k])
        )
```

```python
factor_loadings_df = pd.DataFrame(
    factor_loadings,
    index=[f"Band{k+1}" for k in range(top_3_eigenvectors.shape[0])],
    columns=[f"PC{p+1}" for p in range(top_3_eigenvectors.shape[1])]
)


print("\nTop 3 Factor Loadings (R_{kp}):")
print(factor_loadings_df)
```
