```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D1GB1.csv", header=None, names=['Wavelength', 'Reflectance'])
df2 = pd.read_csv("C:\\Users\\sumat\\Desktop\\GSI Suman\\Processed Files\\D1GB2.csv", header=None, names=['Wavelength', 'Reflectance'])
```

```python
print(df.columns)

# Plotting the data with exact column names
plt.figure(figsize=(10, 6))
plt.plot(df[df.columns[0]], df[df.columns[1]], label='Reflectance')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (%)')
plt.title('Reflectance Spectrum of Crocin 650')
plt.legend()
plt.grid(True)
plt.show()

```

```python
df.columns = [col.strip() for col in df.columns]

# Plotting the data as a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Wvl'], df['Reflect. %'], label='Reflectance', color='red')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (%)')
plt.title('Reflectance Spectrum of Crocin 650')
plt.legend()
plt.grid(True)
plt.show()
```

Plotting both graph in same graph

```python
df1 = pd.read_csv("PSR-1100_SN12B4415_M0633.csv")

df2 = pd.read_csv("PSR-1100_SN12B4415_M0634.csv")

# Check the columns in the DataFrames to ensure correct column names are used
print(df1.columns)
print(df2.columns)

# Plot the data from both CSV files
plt.figure(figsize=(10, 6))

# Plot data from the first file
plt.plot(df1[df1.columns[0]], df1[df1.columns[1]], label='Crocin650: Reflectance', color='blue')

# Plot data from the second file
plt.plot(df2[df2.columns[0]], df2[df2.columns[1]], label='counterfit Durg: Reflectance', color='red')

# Add titles and labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (%)')
plt.title('Reflectance Spectrum Comparison')

# Add a legend
plt.legend()

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Scatter plot data from the first file
plt.scatter(df1[df1.columns[0]], df1[df1.columns[1]], label='Crocin650: Reflectance', color='blue')

# Scatter plot data from the second file
plt.scatter(df2[df2.columns[0]], df2[df2.columns[1]], label='Countefit Drug: Reflectance', color='red')

# Add titles and labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (%)')
plt.title('Reflectance Spectrum Comparison')

# Add a legend
plt.legend()

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()
```

```python
wavelengths = df1[df1.columns[0]]
reflectance_diff = df1[df1.columns[1]] - df2[df2.columns[1]]

# Plot the reflectance spectra from both CSV files and their difference
plt.figure(figsize=(12, 8))

# Scatter plot data from the first file
plt.scatter(df1[df1.columns[0]], df1[df1.columns[1]], label='Crocin650: Reflectance', color='blue')

# Scatter plot data from the second file
plt.scatter(df2[df2.columns[0]], df2[df2.columns[1]], label='Counterfeit Drug: Reflectance', color='red')

# Plot the difference
plt.plot(wavelengths, reflectance_diff, label='Difference in Reflectance', color='green', linestyle='--')

# Add titles and labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (%)')
plt.title('Reflectance Spectrum Comparison')

# Add a legend
plt.legend()

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()
```

how much differnce in %

```python
wavelengths = df1[df1.columns[0]]
reflectance1 = df1[df1.columns[1]]
reflectance2 = df2[df2.columns[1]]
percentage_diff = ((reflectance1 - reflectance2) / reflectance2) * 100

# Plot the reflectance spectra from both CSV files and their percentage difference
plt.figure(figsize=(12, 8))

# Scatter plot data from the first file
plt.scatter(df1[df1.columns[0]], df1[df1.columns[1]], label='Crocin650: Reflectance', color='blue')

# Scatter plot data from the second file
plt.scatter(df2[df2.columns[0]], df2[df2.columns[1]], label='Counterfeit Drug: Reflectance', color='red')

# Plot the percentage difference
plt.plot(wavelengths, percentage_diff, label='Percentage Difference in Reflectance', color='green', linestyle='--')

# Add titles and labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (%) / Percentage Difference (%)')
plt.title('Reflectance Spectrum Comparison and Percentage Difference')

# Add a legend
plt.legend()

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()
```
